```python
import polars as pl
```

```python
mimic3source = "/home/cararc/data/mimic-iii-1.4/"
prescriptions = pl.read_csv(
    mimic3source + "PRESCRIPTIONS.csv",
    separator=",",
    schema_overrides={"GSN": pl.Utf8},
    infer_schema_length=10000
)
```

```python
prescriptions = prescriptions.with_columns(
    pl.struct(["STARTDATE", "DRUG", "GSN", "NDC", "PROD_STRENGTH", "DOSE_VAL_RX", "DOSE_UNIT_RX"]).alias("prescription")
)
prescriptions = prescriptions.group_by(["SUBJECT_ID", "HADM_ID"]).agg(
    pl.col("prescription")
)
# prescriptions = prescriptions.drop(["STARTDATE", "DRUG", "GSN", "NDC", "PROD_STRENGTH", "DOSE_VAL_RX", "DOSE_UNIT_RX"])
```
